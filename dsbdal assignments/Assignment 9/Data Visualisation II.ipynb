```python
df = sns.load_dataset('titanic')
```

```python
# box plot for distribution of age with respect to each gender along with the information about whether they survived 
# or not
```

```python
sns.boxplot(data = df, x = "class", y = "age", hue = "alive" )
```

```python
# in a box plot - the median, the interquatile range and the outliers can be read

# the box contains 50% of data  (this is the interquatile range)
# rest 25-25% lies above and below teh box


# solid line --> median
# dashed line --> mean
```

```python
# box --> lets say lower - q1 and upper q3

# 25% of people have age greater than q3 i.e above 45 
# 25% of people have age less than q1 i.e below around 28
```
